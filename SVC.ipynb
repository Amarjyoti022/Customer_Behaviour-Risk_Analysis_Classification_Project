```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter("ignore")
```

```python
df = pd.read_excel("cleaned_CB.xlsx")
df
```

```python
X = df.drop(columns = ['is_fraudulent'])
y = df['is_fraudulent']
```

**Train_test split**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size = 0.8,random_state = 81)
```

```python
from sklearn.decomposition import PCA
pc = PCA(n_components = 0.95)
X_train = pc.fit_transform(X_train)
X_test = pc.transform(X_test)
```

```python
#Hyper parameter tunning
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

estimator = SVC()
param_grid = {'C':[0.01,0.1,1],'kernel':['linear','rbf','sigmoid','poly']}
svm_grid = GridSearchCV(estimator,param_grid,scoring='accuracy',cv=5)
svm_grid.fit(X_train,y_train)

svm_grid.best_estimator_
```

**Modelling**

```python
from sklearn.svm import SVC
svm_model = SVC(C = 0.01,kernel = 'linear')
svm_model.fit(X_train,y_train)
```

**Prediction**

```python
ytrain_pred = svm_model.predict(X_train)
ytest_pred = svm_model.predict(X_test)
```

**Evaluation**

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score

print('train_accuracy:',accuracy_score(y_train,ytrain_pred))
print('test_accuracy:',accuracy_score(y_test,ytest_pred))
print('cv_score:',cross_val_score(svm_model,X_train,y_train,cv=5,scoring='accuracy').mean())
```
